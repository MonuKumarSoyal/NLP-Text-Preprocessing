```python
from tensorflow.keras.preprocessing.text import one_hot
```

```python
# input sentences

sentences =[  'the glass of milk',
              'the glass of juice',
              'the cup of tea',
             'I am a good boy',
             'I am a good developer',
             'understand the meaning of words',
             'your videos are good',]
```

```python
vac_size = 100
one_hot_repre = [one_hot(words, vac_size) for words in sentences]
one_hot_repre
```

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences
```

```python
max_length = 6
input = pad_sequences(one_hot_repre, padding = "pre", maxlen = max_length)
input
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding
```

```python
model = Sequential()
model.add(Embedding(vac_size, 10, input_length = max_length))
model.compile("adam", "mse")
```

```python
model.summary()
```

```python
print(input[0])
print(model.predict(input[0]))
```

```python
print(model.predict(input))
```
